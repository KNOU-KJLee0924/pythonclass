```python
from math import pi
r = 10
for h in [1,5,14,26,31] :
    print("반지름이", r, "이고, 높이가", h, "인 원뿔의 부피는", (1/3) * pi * r**2 * h, ", 겉넓이는", pi * r**2 + pi * r * h, "입니다.")
```

```python
from math import pi
for rh in [[10,1],[20,5],[30,14]] :
    print("반지름이", rh[0], "이고, 높이가", rh[1], "인 원뿔의 부피는", (1/3) * pi * rh[0]**2 * rh[1], ", 겉넓이는", pi * rh[0]**2 + pi * rh[0] * rh[1], "입니다.")
```

```python
r_list = range(10, 31, 10)
h_list = (1, 5, 14)
for i in range(0,len(r_list))  :
    print("반지름이", r_list[i], "이고, 높이가", h_list[i], "인 원뿔의 부피는", (1/3) * pi * r_list[i]**2 * h_list[i], ", 겉넓이는", pi * r_list[i]**2 + pi * r_list[i] * h_list[i], "입니다.")
```

```python
#zip을 이용하면 동일 인덱스값을 같이 꺼내서 쓸 수 있다
from math import pi
r_list = range(10, 31, 10)
h_list = (1, 5, 14)
for r,h in zip(r_list, h_list) :
    print("반지름이", r, "이고, 높이가", h, "인 원뿔의 부피는", (1/3) * pi * r**2 * h, ", 겉넓이는", pi * r**2 + pi * r * h, "입니다.")
```
