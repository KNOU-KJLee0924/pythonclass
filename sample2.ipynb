```python
# 부피 = 1/3Pir**2h
# 겉넓이 = Pir**2 + Pirh
from math import pi
import math
h = 30
r = 20
Pi = 3.14
print("부피 = " + str(1/3 * Pi * r**2 * h))
print("겉넓이 = " + str(Pi * r**2 + Pi * r * h))
print("")
print("부피 = " + str(1/3 * pi * r**2 * h))
print("겉넓이 = " + str(pi * r**2 + pi * r * h))
print("")
print("부피 = " + str(1/3 * math.pi * r**2 * h))
print("겉넓이 = " + str(math.pi * r**2 + math.pi * r * h))
```
