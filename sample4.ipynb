```python
# if 를 이용한 입력값 확인
from math import pi
r = input("반지름을 입력하세요. :")
if r == "" :
    r = 0
h = input("높이를 입력하세요. :")
if h == "" :
    h = 0
h = float(h)
#부피 volume
if r > 0 and h > 0 :
    v = 1/3 * pi * r**2 * h
    #겉넓이 surface area
    s = pi * r**2 + pi * r * h
    print("입력한 값에 대한 원뿔의 부피는", v ,"이고, 겉넓이는", s,"입니다.")
else :
    print("반지름과 높이는 0보다 커야합니다.")
```

```python
# while 을 이용한 입력값을 필수로 받아서 실행하기
from math import pi
r = ""
h = ""
while r == "" :
    r = input("반지름을 입력하세요. :")
r = float(r)
while h == "" :
    h = input("높이를 입력하세요. :")
h = float(h)
#부피 volume
if r > 0 and h > 0 :
    v = 1/3 * pi * r**2 * h
    #겉넓이 surface area
    s = pi * r**2 + pi * r * h
    print("입력한 값에 대한 원뿔의 부피는", v ,"이고, 겉넓이는", s,"입니다.")
else :
    print("반지름과 높이는 0보다 커야합니다.")
```

```python
a = ""
b = ""
c = ""
while a == "" :
    a = input("첫번째 숫자를 입력하세요. :")
while b == "" :
    b = input("두번째 숫자를 입력하세요. :")
while c == "" :
    c = input("세번째 숫자를 입력하세요. :")


if float(a) > float(b) and float(a) > float(c) :
    print("가장 큰 숫자는", a,"입니다.")
elif float(b) > float(a) and float(b) > float(c) :
    print("가장 큰 숫자는", b,"입니다.")
else :
    print("가장 큰 숫자는", c,"입니다.")
```
