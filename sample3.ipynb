```python
from math import pi
r = input("반지름을 입력하세요. :")
h = input("높이를 입력하세요. :")
print("입력한 값에 대한 원뿔의 부피는 :")
print(1/3 * pi * float(r)**2 * float(h) )
print("이고, 겉넓이는 ")
print( pi * float(r)**2 + pi * float(r) * float(h))
print("입니다. ")
```

```python
from math import pi
r = float(input("반지름을 입력하세요. :"))
h = float(input("높이를 입력하세요. :"))
#부피 volume
v = 1/3 * pi * r**2 * h
#겉넓이 surface area
s = pi * r**2 + pi * r * h
print("입력한 값에 대한 원뿔의 부피는", v ,"이고, 겉넓이는", s,"입니다.")
```
